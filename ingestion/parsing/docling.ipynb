```python
from docling_core.types.doc.document import DoclingDocument, SectionHeaderItem

grouped_sections = []
current_section = []

for node in doc.body.children:
        # Resolve the actual document element (Text, Table, Group, etc.)
        item = node.resolve(doc)
        
        if item is None:
            continue
            
        # Check if the element is an H1, H2, or any other SectionHeaderItem
        if isinstance(item, SectionHeaderItem):
            # Save the previous section if it contains elements
            if current_section:
                grouped_sections.append(current_section)
            # Start a brand new section block, including the header item itself
            current_section = [item]
        else:
            # Append all body text, tables, or structures to the active section
            current_section.append(item)

# Append the final section block after loop completion
if current_section:
    grouped_sections.append(current_section)

grouped_sections[1]
```

```python
from docling_core.types.doc import DoclingDocument, TableItem, TextItem


@dataclass
class ParsedSection:
    section_id: str          # e.g. "section_5_exclusions"
    section_title: str
    page_start: int
    page_end: int
    raw_markdown: str        # Full section as markdown
    tables: list[dict]       # Structured table JSON from Docling
    confidence: float  

def _process_section(items, doc: DoclingDocument) -> ParsedSection:
        """Convert a group of DoclingDocument items into a ParsedSection."""
        tables = []
        text_parts = []
        pages = []
        
        for item in items:
            if item.prov:
                pages.extend([p.page_no for p in item.prov])
            
            if isinstance(item, TableItem):
                # Export table as both markdown and structured JSON
                table_dict = {
                    "markdown": item.export_to_markdown(),
                    "dataframe": item.export_to_dataframe().to_dict(orient="records"),
                    "cells": _extract_table_cells(item),
                    "confidence": getattr(item, "confidence", 1.0),
                }
                tables.append(table_dict)
                text_parts.append(item.export_to_markdown())
            
            elif isinstance(item, TextItem):
                text_parts.append(item.text)
        
        section_markdown = "\n\n".join(text_parts)
        avg_confidence = sum(
            t["confidence"] for t in tables
        ) / len(tables) if tables else 1.0
        
        return ParsedSection(
            section_id=_make_section_id(items[0]),
            section_title=_extract_section_title(items),
            page_start=min(pages) if pages else 0,
            page_end=max(pages) if pages else 0,
            raw_markdown=section_markdown,
            tables=tables,
            confidence=avg_confidence,
        )

def _extract_table_cells(table: TableItem) -> list[dict]:
        """
        Extract cells with parent-label inheritance.
        This handles merged header rows (e.g., "Inpatient Surgery" spanning
        multiple sub-rows like "General", "Cardiac").
        """
        cells = []
        df = table.export_to_dataframe()
        
        # Detect hierarchical rows (parent labels)
        # A row is a "parent" if it has only one non-empty cell in column 0
        # and all other data columns are empty
        parent_label = None
        
        for _, row in df.iterrows():
            row_values = row.tolist()
            non_empty = [v for v in row_values if str(v).strip() not in ("", "nan")]
            
            if len(non_empty) == 1:
                # This is a parent/header row
                parent_label = non_empty[0]
                continue
            
            cell_dict = {
                col: str(val).strip() 
                for col, val in zip(df.columns, row_values)
            }
            if parent_label:
                cell_dict["__parent_label__"] = parent_label
            
            cells.append(cell_dict)
        
        return cells

import hashlib
import re
from docling_core.types.doc.document import SectionHeaderItem

def _extract_section_title(items: list) -> str:
    """Extract the string title from a grouped section list.
    
    Looks for the first SectionHeaderItem in the group. If missing,
    it falls back to a generic 'Untitled Section' label.
    """
    for item in items:
        if isinstance(item, SectionHeaderItem):
            return item.text.strip()
            
    return "Untitled Section"

def _make_section_id(first_item) -> str:
    """Generate a clean, deterministic, and unique ID for a section block.
    
    Uses a slugified version of the header text combined with a short hash 
    of its coordinates or internal sequence pointer to guarantee uniqueness,
    even if multiple sections share identical text names.
    """
    # 1. Fallback text if the first item isn't a header or is empty
    base_text = "section"
    unique_seed = "0"
    
    if first_item is not None:
        if hasattr(first_item, "text") and first_item.text:
            base_text = first_item.text.strip().lower()
        
        # Pull a unique pointer constraint from docling's model schema
        if hasattr(first_item, "prov") and first_item.prov:
            # Uses the page number and element bounding box coordinates as a seed
            unique_seed = str(first_item.prov[0].page_no) + str(first_item.prov[0].bbox)
        elif hasattr(first_item, "self_ref"):
            unique_seed = str(first_item.self_ref)

    # 2. Slugify the text string (keep only alphanumeric characters and hyphens)
    slug = re.sub(r"[^a-z0-9\-_]+", "-", base_text)
    slug = re.sub(r"-+", "-", slug).strip("-")
    slug = slug[:30] if slug else "section" # Truncate long headers cleanly

    # 3. Create a unique 8-character hash suffix
    hash_suffix = hashlib.md5(unique_seed.encode("utf-8")).hexdigest()[:8]

    return f"{slug}-{hash_suffix}"
```
